# Nimber Arithmetic
The main operations are nimber addition and nimber multiplication. These are useful for some impartial games (the winner depends on the position). It can be used in nim game. This notebook is intended for building intuition for the nimber addition and multiplication

## Nimber Addition
Nimber Addition is binary digit summation, which is also the exclusive OR operator 'xor'. Here is a nimber addition table.

![](Cayley_table.svg.png 'nimber addition')

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def nimberAdditionGrid(n):
    matrix = [[0 for _ in range(101)] for _ in range(101)]
    for i in range(n):
        for j in range(n):
            matrix[i][j]=i^j
    fig, ax = plt.subplots(figsize=(n,n))

    min_val, max_val = 0, n

    for i in range(n):
        for j in range(n):
            c = matrix[i][j]
            ax.text(i+0.5,j+0.5, str(c), va='center', ha='center')
    ax.set_xlim(min_val,max_val)
    ax.set_ylim(min_val, max_val)
    ax.set_xticks(np.arange(max_val))
    ax.set_yticks(np.arange(max_val))
    ax.grid()
    fig.savefig('nimberAddition.svg')
```

```python
nimberAdditionGrid(21);
```
